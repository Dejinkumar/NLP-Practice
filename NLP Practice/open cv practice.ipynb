```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img=cv2.imread(r"cv.jpeg",1)
```

```python
plt.imshow(img)
```

```python
nimg = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
```

```python
plt.imshow(nimg)
```

```python
nimg.shape
```

```python

BGRImage = cv2.resize(nimg, (800, 1200))
YCrCbImage = cv2.cvtColor(BGRImage, cv2.COLOR_BGR2YCR_CB)
Y, Cb, Cr = YCrCbImage[:,:,0], YCrCbImage[:,:,1], YCrCbImage[:,:,2]
Y = np.array(Y).astype(np.int16)
Cb = np.array(Cb).astype(np.int16)
Cr = np.array(Cr).astype(np.int16)
Y, Cb, Cr = Y - 128, Cb - 128, Cr - 128
```
